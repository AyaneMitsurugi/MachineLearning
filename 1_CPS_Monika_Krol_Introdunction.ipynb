#### 10. Display histograms for individual parameters with regard to belonging to a given class
##### Get data for a specific flower (versicolor, virginica or setosa) and drop 'target' column out of iris (data frame)

```python
iris_versicolor = iris[iris['target'] == 0.0].drop(columns='target')
iris_virginica  = iris[iris['target'] == 1.0].drop(columns='target')
iris_setosa     = iris[iris['target'] == 2.0].drop(columns='target')
```

**Versicolor**:

```python
plt.figure(figsize=(15,10))
plt.suptitle("VERSICOLOR", fontsize=14)

plt.subplot(2,2,1)
versicolor_sepal_length_hist = plt.hist(iris_versicolor['sepal length (cm)'])
plt.title("sepal length (cm)")

plt.subplot(2,2,2)
versicolor_sepal_width_hist = plt.hist(iris_versicolor['sepal width (cm)'])
plt.title("sepal width (cm)")

plt.subplot(2,2,3)
versicolor_petal_length_hist = plt.hist(iris_versicolor['petal length (cm)'])
plt.title("petal length (cm)")

plt.subplot(2,2,4)
versicolor_petal_width_hist = plt.hist(iris_versicolor['petal width (cm)'])
plt.title("petal width (cm)")
```

**Virginica**:
